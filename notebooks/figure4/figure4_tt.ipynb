```python
import pandas as pd

data = pd.read_excel("r1_phth_timetrial_20240824.xlsx")
```

```python
mat = []
this_mat = []
n = 0
labels = []
this_label = []
for idx,i in data.iterrows():

    this_mat.append(i["P, 4.17"])
    n = n + 1
    this_label.append(i["Unnamed: 0"])
    if n == 4:
        n = 0
        mat.append(this_mat)
        this_mat = []
        labels.append(this_label)
        this_label = []
print(mat)
print(max(max(mat)))
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

def create_blues_colorbar():
    # Create a figure and a single subplot
    fig, ax = plt.subplots(figsize=(0.1, 1.5625))
    
    # Define the colormap
    cmap = mpl.cm.Blues
    
    # Create a normalization from 0 to 1
    norm = mpl.colors.Normalize(vmin=0, vmax=1)
    
    # Create a colorbar with no ticks and no labels
    cbar = mpl.colorbar.ColorbarBase(ax, cmap=cmap,
                                      norm=norm,
                                      orientation='vertical')
    cbar.ax.set_xticks([])  # Remove the ticks
    cbar.ax.set_yticks([])  # Just to be sure even for vertical
    cbar.outline.set_visible(True)  # Hide the colorbar outline


    # Save the figure with high dpi
    fig.savefig('blues_colorbar.png', dpi=600, bbox_inches='tight', pad_inches=0.01)
    plt.close(fig)  # Close the plot to free up memory

# Call the function to create and save the colorbar
create_blues_colorbar()
```
